# MediMine: Exploratory Data Analysis (EDA)
# Goal: Understand the frequency of symptoms before mining.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Processed Data (One-Hot Encoded CSV)
try:
    # We load the CSV we created in the previous notebook
    df = pd.read_csv('../data/processed/encoded_symptoms.csv')
    print(f"✅ Loaded {len(df)} patient records.")
except FileNotFoundError:
    print("❌ Error: Run '01_Data_Preprocessing.ipynb' first.")

# 2. Calculate Symptom Frequency
# Since data is Boolean (True/False), we just sum the columns to get the count
symptom_counts = df.sum().sort_values(ascending=False).head(20)

print("Top 10 Most Frequent Symptoms:")
print(symptom_counts.head(10))

# 3. Visualization: Top 20 Symptoms Bar Chart
plt.figure(figsize=(12, 8))
sns.barplot(x=symptom_counts.values, y=symptom_counts.index, palette="viridis")
plt.title("Top 20 Most Frequent Symptoms in Dataset")
plt.xlabel("Frequency (Count)")
plt.ylabel("Symptom")
plt.grid(axis='x', alpha=0.3)

# Save the plot
plt.tight_layout()
plt.savefig('../results/symptom_distribution.png')
print("✅ EDA Chart saved to '../results/symptom_distribution.png'")
plt.show()
```
